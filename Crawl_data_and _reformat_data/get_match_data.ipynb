```python
## info a team include: 
player = ["lineup_player_id_"+ str(i) for i in range(11)] + ["sub_player_id_" + str(i) for i in range(9)]

feature_team = [ "id_match","season","name", "id", "score", "result", "formation" ] + player + stats
info_team = ["name", "id", "score", "formation" ]

data_team = {}
for i in feature_team:
    data_team[i] = []

print(data_team)
```

```python
for season in seasons:
    # Opening JSON file
    f = open(f'D:/Git_desktop/Intro_data_science_new/Crawl_data_and _reformat_data/match/match/{season}/matches_data.json', encoding="utf8")
     
    # returns JSON object as 
    # a dictionary
    data = json.load(f)

    for datum in data:
        for index in [0,1]:
            print(datum["id"])
            try:
                data_team["id_match"].append(datum["id"])
            except:
                data_team["id_match"].append(-1)
            data_team["season"].append(datum["season"])
            data_team["name"].append(datum["teams"][index]["team"]["name"])
            data_team["id"].append(datum["teams"][index]["team"]["id"])
            data_team["score"].append(datum["teams"][index]["score"])
            if datum["teams"][index]["score"] > datum["teams"][1 - index]["score"]:
                data_team["result"].append("win")
            elif datum["teams"][index]["score"] == datum["teams"][1 - index]["score"]:
                data_team["result"].append("tie")
            else:
                data_team["result"].append("lose")
            try:
                data_team["formation"].append(datum["teamLists"][index]["formation"]["label"])
            except:
                data_team["formation"].append(-1)
    
            # get data lineup_player
            for i in range(11):
                try:
                    data_team["lineup_player_id_" + str(i)].append(datum["teamLists"][index]["lineup"][i]["id"])
                except:
                    data_team["lineup_player_id_" + str(i)].append(-1)
            for i in range(9):
                try:
                    data_team["sub_player_id_" + str(i)].append(datum["teamLists"][index]["substitutes"][i]["id"])
                except:
                    data_team["sub_player_id_" + str(i)].append(-1)
            id_team = str(datum["teams"][index]["team"]["id"])
            for i in stats:
                check = 0
                for stat in datum["stats"][id_team]["M"]:
                    if stat["name"] == i:
                        data_team[i].append(stat["value"])
                        check = 1
                if check == 0:
                    data_team[i].append(-1)
```
