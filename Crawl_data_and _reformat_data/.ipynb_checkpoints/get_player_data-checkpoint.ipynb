```python
data_player = {"id": [], "playerId": [], "name": [], "season": [], "position": [], "shirtNum": [], "positionInfo": [],"country": [], "birth": [], "age": [] }
for i in stats:
    data_player[i] = []
```

```python
for season in seasons:
    # Opening JSON file
    f = open(f'D:/Git_desktop/Intro_data_science/player/player/{season}/players_data.json', encoding="utf8")
     
    # returns JSON object as 
    # a dictionary
    data = json.load(f)

    for datum in data:
        data_player["id"].append(datum["entity"]["id"])
        data_player["playerId"].append(datum["entity"]["playerId"])
        data_player["name"].append(datum["entity"]["name"]["display"])
        data_player["season"].append(datum["season"])
        try:
            data_player["position"].append(datum["entity"]["info"]["position"])
        except:
            data_player["position"].append(-1)
        # print(1)
        if "shirtNum" in datum["entity"]["info"]:
            data_player["shirtNum"].append(datum["entity"]["info"]["shirtNum"])
        else:
            data_player["shirtNum"].append(-1)

        try:
            data_player["positionInfo"].append(datum["entity"]["info"]["positionInfo"])
        except:
            data_player["positionInfo"].append(-1)
            
        if  "country" in datum["entity"]["birth"]["country"] :
            data_player["country"].append(datum["entity"]["birth"]["country"]["country"])
        else:
            data_player["country"].append(-1)

        try:
            data_player["birth"].append(int(datum["entity"]["birth"]["date"]["label"][-4:]))
        except:
            data_player["birth"].append(-1)

        try:
            data_player["age"].append(int(datum["season"][:4]) - int(datum["entity"]["birth"]["date"]["label"][-4:]))
        except:
            data_player["age"].append(-1)
    
        for i in stats:
            if i in datum["stats"]:
                data_player[i].append(datum["stats"][i]["value"])
            else:
                data_player[i].append(-1)
```
