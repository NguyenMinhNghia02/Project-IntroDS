```python
def get_stats(playerId,season):
  list_data = []
  for i in range(len(player_data)):
    if player_data['id'][i] == playerId and player_data['season'][i][0:4] == str(season):
      if player_data['position'][i] == 'G':
        for att in goalkeeper_att:
          list_data.append(player_data[att][i])
        break
      if player_data['position'][i] == 'D':
        for att in defender_att:
          list_data.append(player_data[att][i])
        break
      if player_data['position'][i] == 'M':
        for att in midfielder_att:
          list_data.append(player_data[att][i])
        break
      if player_data['position'][i] == 'F':
        for att in forward_att:
          list_data.append(player_data[att][i])
        break
  return list_data
```

```python
def get_mean_stats(teamId,season):
  G_list=[]
  D_list=[]
  M_list=[]
  F_list=[]
  if season >= 1999:
    season = str(season)+'-'+str(season-1999)
  else:
    season = str(season)+'-'+str(season-1899)
  # print(season)
  id_list = match_season_id['stt'].loc[match_season_id['season'] == season].tolist()[0]
  for id in id_list:
    if match_data['id_1'][id] == teamId:
      for mlu in main_line_up_1:
         playerId = match_data[mlu][id]
         if find_pos(playerId)== 'G' and playerId not in G_list:
          G_list.append(playerId)
         if find_pos(playerId)== 'D' and playerId not in D_list:
          D_list.append(playerId)
         if find_pos(playerId)== 'M' and playerId not in M_list:
          M_list.append(playerId)
         if find_pos(playerId)== 'F' and playerId not in F_list:
          F_list.append(playerId)

    if match_data['id_2'][id] == teamId:
      for mlu in main_line_up_2:
        playerId = match_data[mlu][id]
        # print('Yes')
        if find_pos(playerId)== 'G' and playerId not in G_list:
          G_list.append(playerId)
        if find_pos(playerId)== 'D' and playerId not in D_list:
          D_list.append(playerId)
        if find_pos(playerId)== 'M' and playerId not in M_list:
          M_list.append(playerId)
        if find_pos(playerId)== 'F' and playerId not in F_list:
          F_list.append(playerId)
  # print(G_list)
  # print(D_list)
  # print(M_list)
  # print(F_list)
  G_stats=[]
  D_stats=[]
  M_stats=[]
  F_stats=[]
  season = int(season[0:4])
  for G in G_list:
    if (len(get_stats(G, season)) > 0 and get_stats(G, season).count(-1) < len(get_stats(G, season))):
      glist = get_stats(G, season)
      G_stats.append(glist)
  for D in D_list:
    if (len(get_stats(D, season)) > 0 or get_stats(D, season).count(-1) < len(get_stats(D, season))):
      dlist = get_stats(D, season)
      D_stats.append(dlist)
  for M in M_list:
     if (len(get_stats(M, season)) > 0 or get_stats(M, season).count(-1) < len(get_stats(M, season))):
      mlist = get_stats(M, season)
      M_stats.append(mlist)
  for F in F_list:
     if (len(get_stats(F, season)) > 0 or get_stats(F, season).count(-1) < len(get_stats(F, season))):
      flist = get_stats(F, season)
      F_stats.append(flist)
  # print()
  # print(G_stats)
  # print(D_stats)
  # print(M_stats)
  # print(F_stats)
  arrays = [np.array(x) for x in G_stats]
  G_mean = [np.mean(k) for k in zip(*arrays)]

  arrays = [np.array(x) for x in D_stats]
  D_mean = [np.mean(k) for k in zip(*arrays)]

  arrays = [np.array(x) for x in M_stats]
  M_mean = [np.mean(k) for k in zip(*arrays)]

  arrays = [np.array(x) for x in F_stats]
  F_mean = [np.mean(k) for k in zip(*arrays)]


  return [G_mean, D_mean, M_mean, F_mean]

```

```python
def get_aver_att_season(season):
  G_list=[]
  D_list=[]
  M_list=[]
  F_list=[]
  if season >= 1999:
    if season >= 2009:
      season = str(season)+'-'+str(season-1999)
    else:
      season = str(season)+'-'+ '0'+str(season-1999)
  else:
    season = str(season)+'-'+str(season-1899)
  # print(season)
  id_list = match_season_id['stt'].loc[match_season_id['season'] == season].tolist()[0]
  for id in id_list:

    for mlu in main_line_up_1:
      playerId = match_data[mlu][id]
      pos = find_pos(playerId)
      if pos == 'G' and playerId not in G_list:
        G_list.append(playerId)
      if pos == 'D' and playerId not in D_list:
        D_list.append(playerId)
      if pos == 'M' and playerId not in M_list:
        M_list.append(playerId)
      if pos == 'F' and playerId not in F_list:
        F_list.append(playerId)


    for mlu in main_line_up_2:
      playerId = match_data[mlu][id]
      pos = find_pos(playerId)
      if pos == 'G' and playerId not in G_list:
        G_list.append(playerId)
      if pos == 'D' and playerId not in D_list:
        D_list.append(playerId)
      if pos == 'M' and playerId not in M_list:
        M_list.append(playerId)
      if pos == 'F' and playerId not in F_list:
        F_list.append(playerId)
  # print(G_list)
  # print(D_list)
  # print(M_list)
  # print(F_list)
  G_stats=[]
  D_stats=[]
  M_stats=[]
  F_stats=[]
  season = int(season[0:4])
  for G in G_list:
    stats = get_stats(G,season)
    if (len(stats) > 0 and stats.count(-1) < len(stats)):
      glist = get_stats(G, season)
      G_stats.append(glist)
  for D in D_list:
    stats = get_stats(D,season)
    if (len(stats) > 0 and stats.count(-1) < len(stats)):
      dlist = get_stats(D, season)
      D_stats.append(dlist)
  for M in M_list:
    stats = get_stats(M,season)
    if (len(stats) > 0 and stats.count(-1) < len(stats)):
      mlist = get_stats(M, season)
      M_stats.append(mlist)
  for F in F_list:
    stats = get_stats(F,season)
    if (len(stats) > 0 and stats.count(-1) < len(stats)):
      flist = get_stats(F, season)
      F_stats.append(flist)
  # print()
  # print(G_stats)
  # print(D_stats)
  # print(M_stats)
  # print(F_stats)
  arrays = [np.array(x) for x in G_stats]
  G_mean = [np.mean(k) for k in zip(*arrays)]

  arrays = [np.array(x) for x in D_stats]
  D_mean = [np.mean(k) for k in zip(*arrays)]

  arrays = [np.array(x) for x in M_stats]
  M_mean = [np.mean(k) for k in zip(*arrays)]

  arrays = [np.array(x) for x in F_stats]
  F_mean = [np.mean(k) for k in zip(*arrays)]


  return [G_mean, D_mean, M_mean, F_mean]
```

```python
def find_nearest_stats(playerId, season):
  count = 0
  stat = get_stats(playerId, season)
  if season >= 1993:
    while (len(stat) == 0 or stat.count(-1) == len(stat)) and count <= 1:
      season -= 1
      count+= 1
      stat = get_stats(playerId, season)
    return stat
  else:
    return stat
```

```python
def find_inf_2_teams(matchId):
  team1 = []
  team2 = []
  teamid1 = 0
  teamid2 = 0
  for i in range(len(match_data)):
    if int(match_data['id'][i]) == matchId:
      season = int(match_data['season'][i][0:4])
      team1 = team_1[i]
      teamid1 = match_data['id_1'][i]
      team2 = team_2[i]
      teamid2 = match_data['id_2'][i]
      return [season, teamid1, team1, teamid2, team2]
```

```python
def find_att_each_team(matchId):
  season, teamid1, team1, teamid2, team2 = find_inf_2_teams(matchId)
  # print(season, teamid1, team1, teamid2, team2)
  G_list_1=[]
  D_list_1=[]
  M_list_1=[]
  F_list_1=[]

  G_stats_1=[]
  D_stats_1=[]
  M_stats_1=[]
  F_stats_1=[]

  aver_ss = []
  aver_pre_ss = []
  for i in range(len(X)):
    if season == X[i][0]:
      aver_ss = X[i][1]
    if season-1 == X[i][0]:
      aver_pre_ss = X[i][1]
  if season >= 1993:
    G_mean_1, D_mean_1, M_mean_1, F_mean_1 = aver_pre_ss
  else:
    G_mean_1, D_mean_1, M_mean_1, F_mean_1 = aver_ss
  for ply_id in team1:
    pos = find_pos(ply_id)
    if pos == 'G':
      G_list_1.append(ply_id)
    if pos == 'D':
      D_list_1.append(ply_id)
    if pos == 'M':
      M_list_1.append(ply_id)
    if pos == 'F':
      F_list_1.append(ply_id)
    stats = find_nearest_stats(ply_id, season - 1)
    if len(stats) == 0 or stats.count(-1) == len(stats):
      # print(ply_id)
      if pos == 'G':
        G_stats_1.append(G_mean_1)
      if pos == 'D':
        D_stats_1.append(D_mean_1)
      if pos == 'M':
        M_stats_1.append(M_mean_1)
      if pos == 'F':
        F_stats_1.append(F_mean_1)
    else :
      if pos == 'G':
          G_stats_1.append(stats)
      if pos == 'D':
        D_stats_1.append(stats)
      if pos == 'M':
        M_stats_1.append(stats)
      if pos == 'F':
        F_stats_1.append(stats)
    # print(ply_id, find_pos(ply_id), stats)
  # print()
  arrays = [np.array(x) for x in G_stats_1]
  G_mean_1 = [np.mean(k) for k in zip(*arrays)]

  arrays = [np.array(x) for x in D_stats_1]
  D_mean_1 = [np.mean(k) for k in zip(*arrays)]

  arrays = [np.array(x) for x in M_stats_1]
  M_mean_1 = [np.mean(k) for k in zip(*arrays)]

  arrays = [np.array(x) for x in F_stats_1]
  F_mean_1 = [np.mean(k) for k in zip(*arrays)]


  G_list_2=[]
  D_list_2=[]
  M_list_2=[]
  F_list_2=[]

  G_stats_2=[]
  D_stats_2=[]
  M_stats_2=[]
  F_stats_2=[]
  if season >= 1993:
    G_mean_2, D_mean_2, M_mean_2, F_mean_2 = aver_pre_ss
  else:
    G_mean_2, D_mean_2, M_mean_2, F_mean_2 = aver_ss
  for ply_id in team2:
    pos = find_pos(ply_id)
    if pos == 'G':
      G_list_2.append(ply_id)
    if pos == 'D':
      D_list_2.append(ply_id)
    if pos == 'M':
      M_list_2.append(ply_id)
    if pos == 'F':
      F_list_2.append(ply_id)
    stats = find_nearest_stats(ply_id, season - 1)
    if len(stats) == 0 or stats.count(-1) == len(stats):
      # print(ply_id)
      if pos == 'G':
        G_stats_2.append(G_mean_2)
      if pos == 'D':
        D_stats_2.append(D_mean_2)
      if pos == 'M':
        M_stats_2.append(M_mean_2)
      if pos == 'F':
        F_stats_2.append(F_mean_2)
    else :
      if pos == 'G':
        G_stats_2.append(stats)
      if pos == 'D':
        D_stats_2.append(stats)
      if pos == 'M':
        M_stats_2.append(stats)
      if pos == 'F':
        F_stats_2.append(stats)
    # print(ply_id, find_pos(ply_id), stats)

  arrays = [np.array(x) for x in G_stats_2]
  G_mean_2 = [np.mean(k) for k in zip(*arrays)]

  arrays = [np.array(x) for x in D_stats_2]
  D_mean_2 = [np.mean(k) for k in zip(*arrays)]

  arrays = [np.array(x) for x in M_stats_2]
  M_mean_2 = [np.mean(k) for k in zip(*arrays)]

  arrays = [np.array(x) for x in F_stats_2]
  F_mean_2 = [np.mean(k) for k in zip(*arrays)]


  return[G_mean_1, D_mean_1, M_mean_1, F_mean_1, G_mean_2, D_mean_2, M_mean_2, F_mean_2]
```
